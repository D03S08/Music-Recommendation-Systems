```python
from google.colab import drive
drive.mount('./MyDrive')

import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Membaca dataset
df = pd.read_csv('/content/MyDrive/MyDrive/songs.csv')
df
```

```python
# Membuat matriks peringkat
ratings = pd.pivot_table(df, values='Popularity', index='Name', columns='Artist', fill_value=0)
ratings
```

```python
# Menghitung kemiripan antara pengguna menggunakan cosine similarity
user_similarity = cosine_similarity(ratings)

# Fungsi untuk mendapatkan rekomendasi berdasarkan pengguna yang mirip
def get_recommendations(user):
    similar_users = user_similarity[user].argsort()[:-1][::-1]
    recommendations = []

    for other_user in similar_users:
        similar_user_ratings = ratings.iloc[other_user]
        user_ratings = ratings.iloc[user]

        # Menemukan item yang belum dilihat oleh pengguna
        unseen_items = similar_user_ratings[user_ratings == 0].index

        # Menambahkan item yang belum dilihat oleh pengguna ke rekomendasi
        recommendations.extend(unseen_items)

    return list(set(recommendations))
```

```python
# Contoh pengguna yang akan mendapatkan rekomendasi
user_to_recommend = 10  # Pilih nilai indeks pengguna yang ingin dilihat rekomendasinya

# Mendapatkan rekomendasi untuk pengguna
recommendations = get_recommendations(user_to_recommend)

# Menampilkan rekomendasi
print(f"Rekomendasi untuk pengguna {user_to_recommend}:")
for recommendation in recommendations:
    print(f"- {recommendation}")
```
